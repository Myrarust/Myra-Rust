```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in the data
df = pd.read_csv("airline-safety.csv")
df.head()
```

```python
# Rename columns to prepare to use pd.wide_to_long funtion. Need stubnames and numerical suffixes.
df2 = df.rename(columns={"airline": "airline", "avail_seat_km_per_week": "seats_per_week",
                         "incidents_85_99": "incidents1999","fatal_accidents_85_99": "fatalAccidents1999",
                         "fatalities_85_99": "fatalities1999", "incidents_00_14": "incidents2014", 
                         "fatal_accidents_00_14": "fatalAccidents2014", "fatalities_00_14": "fatalities2014"})
df2.head()
```

```python
df3 = pd.wide_to_long(df2, ["incidents", "fatalAccidents", "fatalities"], i="airline", j="year")
df3.head()
```

```python
# Reset the index after changing format
df3 = df3.reset_index()
```

```python
# Change the format of year 
df4 = df3.replace({'year': {1999: "1985-1999", 2014: "2000-2014"}})
df4.head()
```

```python
df4.airline = df4.airline.replace('*', '')
df4.head()
```

```python
# Change seats per week column from kilometers to miles
df4["seats_per_week"] = df4["seats_per_week"] * 0.62137
```

```python
# Create other statistical column to compare to motor vehicle statistics
df4["airline_deaths_by_100mSeats"] = df4["fatalities"]/df4["seats_per_week"]*100000
df4.head()
```

```python
# Save this dataframe to work with in Tableau
df4.to_excel('airline_clean.xlsx') 
```

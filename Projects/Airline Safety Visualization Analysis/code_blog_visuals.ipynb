```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in the data
df = pd.read_excel("total_miles_by_year.xls")
df.head()
```

```python
# Don't need the first three columns for my viz, drop them. 
df.drop(columns=['Country Code', "Indicator Name"], inplace=True)
```

```python
df
```

```python
# I see a lot of NaNs in the miles columns. Replace those with a value of 0.
df.fillna(0, inplace=True)
```

```python
# I want year to be a column, so I'm going to transpose the dataframe and reset the index.
df = df.set_index('Country Name').T.rename_axis('Year')
df.reset_index(inplace=True)
df
```

```python
# Sweet! I just realized there is a column for world, which is what I need, so I'm going to subset year and world columns. 
world_miles = df[['Year', 'World']]
world_miles
```

```python
# Write to csv to use in Tableau
world_miles.to_csv('world_miles.csv', index=False)
```
